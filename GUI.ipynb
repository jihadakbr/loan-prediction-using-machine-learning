```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
from PIL import ImageTk, Image
import numpy as np
import pandas as pd
import joblib

root = Tk()
logo = Image.open("github.png")
logo = logo.resize((50, 50))
logo = ImageTk.PhotoImage(logo)
logo_label = Label(image=logo)
logo_label.image = logo
logo_label.grid(row=0, column=0, sticky='w', padx=(40,0),pady=30)
Label(root,text = "jihadakbr", font='Courier 16').grid(row=0,column=0,padx=(30,0),pady=30)

root.title('Loan Prediction')
Label(root,text = "Loan Prediction\nUsing Machine Learning", font='Raleway 18 bold').grid(row=0,column=1,pady=30)
Label(root,text = "").grid(row=0,column=2,pady=30)

canvas = Canvas(root, width=800, height=300)
canvas.grid(columnspan=3,rowspan=12)

city_c = pd.read_csv('city_c.csv')
prof_c = pd.read_csv('prof_c.csv')
state_c = pd.read_csv('state_c.csv')

city_c_name = city_c['City'].values.tolist()
city_c_en = city_c['Encoded City'].values.tolist()

prof_c_name = prof_c['Profession'].values.tolist()
prof_c_en = prof_c['Encoded Profession'].values.tolist()

state_c_name = state_c['STATE'].values.tolist()
state_c_en = state_c['Encoded STATE'].values.tolist()

def predict_click():
    t1 = income.get()
    t2 = age.get()
    t3 = exp.get()
    t4 = ms.get()
    t5 = ho.get()
    t6 = co.get()
    t7 = prof.get()
    t8 = city.get()
    t9 = state.get()    
    t10 = crjob.get()
    t11 = crhouse.get()
    
    try:
        int(t1)
        int(t2)
        int(t3)
        int(t10)
        int(t11)
        
        if  t4 == "select one" or t5 == "select one" or t6 == "select one" or t7 == "select one" or t8 == "select one" or t9 == "select one":
            messagebox.showerror('Error', 'Select an option')
        else:
            t7 = prof_com.current()
            t8 = city_com.current()
            t9 = state_com.current()
            
            if ms_com.get() == "Married":
                t4 = 1
            else:
                t4 = 0

            if ho_com.get() == "Owned":
                t5 = 1
            elif ho_com.get() == "Rented":
                t5 = 0
            else:
                t5 = 2

            if co_com.get() == "Yes":
                t6 = 1
            else:
                t6 = 0            
    except ValueError:
        messagebox.showerror('Error', 'Enter a valid number')

    p1 = float(t1)
    p2 = float(t2)
    p3 = float(t3)
    p4 = float(t4)
    p5 = float(t5)
    p6 = float(t6)
    p7 = float(t7)
    p8 = float(t8)
    p9 = float(t9)
    p10 = float(t10)
    p11 = float(t11)

    ET = joblib.load('ET_model')
    scaler = joblib.load('scaler_ET')
    
    input_data = [p1,p2,p3,p4,p5,p6,p7,p8,p9,p10,p11]
    
    input_data_as_numpy_array = np.asarray(input_data)
    input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
    std_data = scaler.transform(input_data_reshaped)
    prediction = ET.predict(std_data)    
        
    if prediction[0] == 0:
        result.configure(text="Non-Defaulter", bg="yellow", fg="red")
    else:
        result.configure(text="Defaulter", bg="yellow", fg="red")

def num1(event):
    try:
        float(e1.get())
        num_e1.config(text="Valid Number",fg="blue")
    except ValueError:
        num_e1.config(text="Invalid Number",fg="red")
def num2(event):
    try:
        float(e2.get())
        num_e2.config(text="Valid Number",fg="blue")
    except ValueError:
        num_e2.config(text="Invalid Number",fg="red")
def num3(event):
    try:
        float(e3.get())
        num_e3.config(text="Valid Number",fg="blue")
    except ValueError:
        num_e3.config(text="Invalid Number",fg="red")
def num4(event):
    try:
        float(e4.get())
        num_e4.config(text="Valid Number",fg="blue")
    except ValueError:
        num_e4.config(text="Invalid Number",fg="red")
def num5(event):
    try:
        float(e5.get())
        num_e5.config(text="Valid Number",fg="blue")
    except ValueError:
        num_e5.config(text="Invalid Number",fg="red")
        
def click1(*args):
    e1.delete(0, 'end')
def click2(*args):
    e2.delete(0, 'end')
def click3(*args):
    e3.delete(0, 'end')    
def click4(*args):
    e4.delete(0, 'end')
def click5(*args):
    e5.delete(0, 'end')  


#### String Var ###    
income = StringVar(root)
age = StringVar(root)
exp = StringVar(root)
crjob = StringVar(root)
crhouse = StringVar(root)

########## column 0 ##########
        
# Income, Age, Experience, Married/Single
e1 = Entry(root, width=40, textvariable=income)
e2 = Entry(root, width=40, textvariable=age)
e3 = Entry(root, width=40, textvariable=exp)

### Income ###
e1.insert(0, 'type here ...')
e1.bind("<Button-1>", click1)
e1.bind("<Motion>", num1)
num_e1 = Label(root, text='')

Label(root, text='Income ($)').grid(row=1,column=0, sticky='w',padx=(40,0),pady=(0,5))
e1.grid(row=2,column=0, sticky='w',padx=(40,0),ipady=10)
num_e1.grid(row=3,column=0, sticky='w',padx=(40,0),pady=(5,25))

### Age ###
e2.insert(0, 'type here ...')
e2.bind("<Button-1>", click2)
e2.bind("<Motion>", num2)
num_e2 = Label(root, text='')

Label(root, text='Age (years)').grid(row=4,column=0, sticky='w',padx=(40,0),pady=(0,5))
e2.grid(row=5,column=0, sticky='w',padx=(40,0),ipady=10)
num_e2.grid(row=6,column=0, sticky='w',padx=(40,0),pady=(5,25))

### Experience ###
e3.insert(0, 'type here ...')
e3.bind("<Button-1>", click3)
e3.bind("<Motion>", num3)
num_e3 = Label(root, text='')

Label(root, text='Professional experience (years): type here').grid(row=7,column=0, sticky='w',padx=(40,0),pady=(0,5))
e3.grid(row=8,column=0, sticky='w',padx=(40,0),ipady=10)
num_e3.grid(row=9,column=0, sticky='w',padx=(40,0),pady=(5,25))

### Married/Single ###
ms = StringVar(root)
ms.set("select one")
ms_com = ttk.Combobox(root, textvariable=ms, value=['Married','Single'], width=37, state="readonly")
ms_com.bind("<<ComboboxSelected>>", predict_click)

Label(root, text='Married/Single').grid(row=10,column=0, sticky='w',padx=(40,0),pady=(0,5))
ms_com.grid(row=11,column=0, sticky='w',padx=(40,0),ipady=10,pady=(0,50))


########## column 1 ##########

### House_Ownership ###
ho = StringVar(root)
ho.set("select one")
ho_com = ttk.Combobox(root, textvariable=ho, value=['Owned','Rented','Neither'], width=37, state="readonly")
ho_com.bind("<<ComboboxSelected>>", predict_click)

Label(root, text='House Ownership').grid(row=1,column=1, sticky='w',padx=(40,0),pady=(0,5))
ho_com.grid(row=2,column=1,padx=40,ipady=10)
Label(root, text='').grid(row=3,column=1,padx=40,pady=(5,25))

### Car_Ownership ###
co = StringVar(root)
co.set("select one")
co_com = ttk.Combobox(root, textvariable=co, value=['Yes','No'], width=37, state="readonly")
co_com.bind("<<ComboboxSelected>>", predict_click)

Label(root, text='Car Ownership').grid(row=4,column=1, sticky='w',padx=(40,0),pady=(0,5))
co_com.grid(row=5,column=1,padx=40,ipady=10)
Label(root, text='').grid(row=6,column=1,padx=40,pady=(5,25))

### Profession ###
prof = StringVar(root)
prof.set("select one")
prof_com = ttk.Combobox(root, textvariable=prof, value=prof_c_name, width=37, state="readonly")
prof_com.bind("<<ComboboxSelected>>", predict_click)

Label(root, text='Profession').grid(row=7,column=1, sticky='w',padx=(40,0),pady=(0,5))
prof_com.grid(row=8,column=1,padx=40,ipady=10)
Label(root, text='').grid(row=9,column=1,padx=40,pady=(5,25))
    
### City ###
city = StringVar(root)
city.set("select one")
city_com = ttk.Combobox(root, textvariable=city, value=city_c_name, width=37, state="readonly")
city_com.bind("<<ComboboxSelected>>", predict_click)

Label(root, text='City').grid(row=10,column=1, sticky='w',padx=(40,0),pady=(0,5))
city_com.grid(row=11,column=1,padx=40,ipady=10,pady=(0,50))


########## column 2 ##########

### State ###
state = StringVar(root)
state.set("select one")
state_com = ttk.Combobox(root, textvariable=state, value=state_c_name, width=37, state="readonly")
state_com.bind("<<ComboboxSelected>>", predict_click)

Label(root, text='State').grid(row=1,column=2, sticky='w',padx=(0,40),pady=(0,5))
state_com.grid(row=2,column=2, sticky='e',padx=(0,40),ipady=10)
Label(root, text='').grid(row=3,column=2,sticky='e',padx=(0,40),pady=(5,25))

### Current Job (years) ###
e4 = Entry(root, width=40,textvariable=crjob)
e5 = Entry(root, width=40,textvariable=crhouse)

e4.insert(0, 'type here ...')
e4.bind("<Button-1>", click4)
e4.bind("<Motion>", num4)
num_e4 = Label(root, text='')

Label(root, text='Current Job (years)').grid(row=4,column=2, sticky='w',padx=(0,40),pady=(0,5))
e4.grid(row=5,column=2, sticky='e',padx=(0,40),ipady=10)
num_e4.grid(row=6,column=2,sticky='w',padx=(0,40),pady=(5,25))

### Current House (years) ###
e5.insert(0, 'type here ...')
e5.bind("<Button-1>", click5)
e5.bind("<Motion>", num5)
num_e5 = Label(root, text='')

Label(root, text='Current House (years)').grid(row=7,column=2, sticky='w',padx=(0,40),pady=(0,5))
e5.grid(row=8,column=2, sticky='e',padx=(0,40),ipady=10)
num_e5.grid(row=9,column=2,sticky='w',padx=(0,40),pady=(5,25))

Label(root, text='').grid(row=10,column=2,sticky='e',padx=(0,40),pady=(0,5))


########## Prediction Button ##########
button = Button(root,text="Predict",command=predict_click,bg="#20bebe", fg="white", font="Raleway 10 bold",
       height=2, width=12)
button.grid(row=11,column=2,sticky='w',pady=(0,50))


########## Result ##########
result = Label(root, text="Result",relief="sunken", fg="#666666", bg="#cccccc",font="Raleway 10",height=2, width=12)
result.grid(row=11, column=2,ipady=4,sticky='e',padx=(0,40),pady=(0,50))

root.mainloop()
```
